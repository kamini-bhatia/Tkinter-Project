```python
import tkinter as tk

def is_prime(num: int) -> bool:
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def check_prime():
    n_str = num_entry.get()

    if not n_str.isdigit():
        result_label.config(text="Please enter a valid positive integer.")
        return

    n = int(n_str)
    if is_prime(n):
        result_label.config(text=f"{n} is a PRIME number.")
    else:
        result_label.config(text=f"{n} is NOT a prime number.")

# ==== MAIN WINDOW ====
root = tk.Tk()
root.title("Prime Number Checker")
root.geometry("300x180")

title_label = tk.Label(root, text="Prime Number Checker", font=("Arial", 14, "bold"))
title_label.pack(pady=5)

frame = tk.Frame(root)
frame.pack(pady=10)

tk.Label(frame, text="Enter a number:").grid(row=0, column=0, padx=5)
num_entry = tk.Entry(frame, width=10)
num_entry.grid(row=0, column=1, padx=5)

check_btn = tk.Button(root, text="Check Prime", command=check_prime)
check_btn.pack(pady=5)

result_label = tk.Label(root, text="")
result_label.pack(pady=5)

root.mainloop()
```
