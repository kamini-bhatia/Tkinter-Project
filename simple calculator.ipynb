```python
# simple calculator
import tkinter as tk

# ----------------- FUNCTIONS -------------------
def button_click(value):
    current = entry_box.get()
    entry_box.delete(0, tk.END)
    entry_box.insert(0, current + value)

def clear():
    entry_box.delete(0, tk.END)

def calculate():
    try:
        expression = entry_box.get()
        result = eval(expression)   # evaluates string expression
        entry_box.delete(0, tk.END)
        entry_box.insert(0, str(result))
    except:
        entry_box.delete(0, tk.END)
        entry_box.insert(0, "Error")

# ----------------- WINDOW -----------------------
root = tk.Tk()
root.title("Simple Calculator")
root.geometry("300x400")

# ----------------- ENTRY BOX --------------------
entry_box = tk.Entry(root, font=("Arial", 20), bd=5, justify="right")
entry_box.pack(fill="both", pady=10, padx=10)

# ------------- FRAME FOR BUTTONS ----------------
frame = tk.Frame(root)
frame.pack()

# All calculator buttons in a list (row-wise)
buttons = [
    "7", "8", "9", "/",
    "4", "5", "6", "*",
    "1", "2", "3", "-",
    "0", "C", "=", "+"
]

row = 0
col = 0

# ----------------- CREATE BUTTONS ---------------
for button in buttons:
    action = lambda x=button: button_click(x)

    if button == "C":
        tk.Button(frame, text=button, width=6, height=2, bg="red",
                  command=clear).grid(row=row, column=col, padx=5, pady=5)

    elif button == "=":
        tk.Button(frame, text=button, width=6, height=2, bg="lightgreen",
                  command=calculate).grid(row=row, column=col, padx=5, pady=5)

    else:
        tk.Button(frame, text=button, width=6, height=2,
                  command=action).grid(row=row, column=col, padx=5, pady=5)

    col += 1
    if col > 3:
        col = 0
        row += 1

# ----------------- START WINDOW ------------------
root.mainloop()
```
